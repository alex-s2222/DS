## 4.1) Провести анализ показателей, полученных с сайта

```python
import os
import requests
import opendatasets as od
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class Kaggle:
    kaggle_path = "https://www.kaggle.com/datasets/lainguyn123/student-performance-factors/data"
    file_name = "StudentPerformanceFactors.csv"
    
    @classmethod
    def __download_file(cls) -> bool:
        try:
            od.download(cls.kaggle_path, f"{os.getcwd()}/")
            print("Download succsess")
            return True
        except Exception as e:
            print(f"Error {e}")
            return False
    
    @classmethod
    def __read_df(cls) -> pd.DataFrame:
        df_name = cls.kaggle_path.split('/')[-2]
        file_path = f"{df_name}/{cls.file_name}"
        
        df = pd.read_csv(file_path)
        return df
    
    @classmethod
    def download_and_get(cls) -> pd.DataFrame | None:
        if cls.__download_file():
            return cls.__read_df()
        print("error")
```

```python
net_df = Kaggle.download_and_get()
```

```python
net_df.head()
```

```python
exam_score = net_df['Exam_Score'].value_counts().sort_index()
plt.figure(figsize=(10,7))
exam_score.plot(kind='bar', color='teal')

for i, value in enumerate(exam_score):
    plt.text(i, value + 6, str(int(value)), ha='center', )
    

plt.title('Результаты экзаменов')
plt.xlabel('Кол-во балов')
plt.ylabel('Кол-во')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##  Практическая часть: Создать статистику по выбранным из таблицы данным в виде бинарного вектора и в виде графиков. Создать объединенные графики статистики.


```python
# Выбираем только числовые столбцы для корреляции
numeric_df = net_df.select_dtypes(include=['int64'])

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Корреляционная тепловая карта числовых признаков')
plt.tight_layout()
plt.show()
```

```python
features = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Tutoring_Sessions']

fig, axs = plt.subplots(len(features), 1, figsize=(10, 18))
for i, feature in enumerate(features):
    sns.boxplot(data=net_df, x=feature, y='Exam_Score', ax=axs[i])
    axs[i].set_title(f'Exam_Score в зависимости от {feature}')
plt.tight_layout()
plt.show()
```

```python
# Преобразуем категориальные признаки в бинарные векторы с помощью one-hot encoding
categorical_df = net_df.select_dtypes(include=['object'])
encoded_df = pd.get_dummies(categorical_df, drop_first=True)

# Объединяем обратно с числовыми признаками
final_df = pd.concat([net_df.select_dtypes(include=['int64']), encoded_df], axis=1)

final_df.head()
```

```python
selected = ['Hours_Studied', 'Attendance', 'Previous_Scores', 'Tutoring_Sessions', 'Exam_Score']

correlations = final_df.corr()['Exam_Score'].drop('Exam_Score').sort_values(key=abs, ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=correlations.values, y=correlations.index, palette='viridis')
plt.title('Топ-10 признаков по абсолютной корреляции с Exam_Score')
plt.xlabel('Коэффициент корреляции')
plt.tight_layout()
plt.show()
```
